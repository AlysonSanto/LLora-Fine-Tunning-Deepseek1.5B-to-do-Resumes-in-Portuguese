```python
from datasets import load_dataset, load_from_disk

dataset = load_from_disk("/content/CV_dataset_local")
print(dataset)
print(dataset[0])
```

Loading the Model

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Defina o identificador do modelo
model_id = "deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B"

# Carregue o tokenizer e o modelo
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_id, trust_remote_code=True)
```

```python
# Criei a instância device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model.to(device)
```

Tokenizer

```python
def tokenize_function(example):
    tokenized = tokenizer(
        example["text"],
        padding="max_length",
        truncation=True,
        max_length=1024,
        return_tensors="pt"
    )
    tokenized["labels"] = tokenized["input_ids"].clone()
    return tokenized

tokenized_dataset = dataset.map(tokenize_function, batched=True)
tokenized_dataset.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
```

LLora

Printing Layers

```python
from peft import get_peft_model, LoraConfig, TaskType


# LoRa
lora_config = LoraConfig(
    r=4,  # rank pequeno
    lora_alpha=16,
    lora_dropout=0.05,
    target_modules=["q_proj", "v_proj"],  # camadas onde aplicar LoRA
    task_type=TaskType.CAUSAL_LM
)

# 3. Adicionando LoRa
model = get_peft_model(model, lora_config)


model.print_trainable_parameters()
```

```python
for name, param in model.named_parameters():
    if param.requires_grad:
        print(name)
```

Dataloader

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(
    tokenized_dataset,
    batch_size=1,  # Ajustar, Nvidia T4 tem em torno de 16GB de VRAM em CPU no mínimo 64GB de RAM
    shuffle=True
)
```

Otimizator Function e Loss

```python
import torch
from torch import nn
from transformers import get_scheduler
num_epochs = 2  # ou qualquer número de épocas desejadas

optimizer = torch.optim.AdamW(model.parameters(), lr=2e-4)
lr_scheduler = get_scheduler(
    "linear",
    optimizer=optimizer,
    num_warmup_steps=100,
    num_training_steps=len(train_dataloader) * num_epochs
)
```

Training Loop

```python
model.train()
for epoch in range(num_epochs):
    print(f"Epoch {epoch+1}/{num_epochs}")
    for batch in train_dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}

        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()

        print(f"Loss: {loss.item():.4f}")
```
